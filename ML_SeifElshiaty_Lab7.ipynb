# Loading the data

First, let's load the train/test sets and take a look at the data.

```python
import pandas as pd
```

```python
train = pd.read_csv('reviews_train.csv')
test = pd.read_csv('reviews_test.csv')

test.sample(5)
```

# Training a baseline model

There are many approaches for training a sequence classification model for text data. In this lab, we're giving you code that mirrors what you find if you look up [how to train a text classifier](https://scikit-learn.org/stable/tutorial/text_analytics/working_with_text_data.html), where we'll train an SVM on [tf-idf](https://en.wikipedia.org/wiki/Tf%E2%80%93idf) features (numeric representations of each text field based on word occurrences).

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
```

```python
sgd_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', SGDClassifier()),
])
```

```python
_ = sgd_clf.fit(train['review'], train['label'])
```

## Evaluating model accuracy

```python
from sklearn import metrics
```

```python
def evaluate(clf):
    pred = clf.predict(test['review'])
    acc = metrics.accuracy_score(test['label'], pred)
    print(f'Accuracy: {100*acc:.1f}%')
```

```python
evaluate(sgd_clf)
```

### 🔵 Tuning Linear SVM

To test whether another classifier can outperform the tuned Multinomial Naive Bayes model, I will now tune the hyperparameters of **Linear SVM** while keeping the same preprocessing pipeline structure unchanged.

```python
from sklearn.svm import LinearSVC
```

```python
svm_clf_tuned = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', LinearSVC(C=0.01, max_iter=5000)),
])
```

### 🔵 Tuning Multinomial Naive Bayes (Second Attempt)

Since the default Multinomial Naive Bayes model performed well, I applied further tuning to both the preprocessing steps and classifier settings in order to improve accuracy.

In this experiment:
- **ngram_range=(1,2)** allows the model to learn both single words and two-word phrases.
- **use_idf=False** removes inverse document frequency weighting.
- **sublinear_tf=True** scales repeated word counts more smoothly.
- **alpha=5.0** adjusts the smoothing parameter of the Naive Bayes classifier.

This remains a model-centric approach because the dataset is unchanged while the model configuration is improved.

```python
nb_clf_tuned = Pipeline([
    ('vect', CountVectorizer(ngram_range=(1,2))),
    ('tfidf', TfidfTransformer(use_idf=False, sublinear_tf=True)),
    ('clf', MultinomialNB(alpha=5.0)),
])

_ = nb_clf_tuned.fit(train['review'], train['label'])

evaluate(nb_clf_tuned)
```

### Model Evaluation

The tuned Multinomial Naive Bayes model achieved an accuracy of **90.5%**, which improved upon the original Multinomial Naive Bayes result of **85.3%**.

This improvement suggests that using phrase-based features and adjusting model settings helped the classifier learn more useful sentiment patterns from the reviews.

### 🔵 Further Tuning Multinomial Naive Bayes (Third Attempt)

To achieve even better performance, an additional tuning experiment was performed by increasing the smoothing parameter `alpha` while keeping the improved preprocessing settings unchanged.

This experiment tests whether stronger smoothing can improve generalization on a noisy text dataset.

```python
nb_clf_tuned2 = Pipeline([
    ('vect', CountVectorizer(ngram_range=(1,2))),
    ('tfidf', TfidfTransformer(use_idf=False, sublinear_tf=True)),
    ('clf', MultinomialNB(alpha=26.0)),
])

_ = nb_clf_tuned2.fit(train['review'], train['label'])

evaluate(nb_clf_tuned2)
```

### Model Evaluation

Using `alpha = 26.0`, the Multinomial Naive Bayes model achieved an accuracy of **91.3%**, which is the highest result obtained so far.

This indicates that stronger smoothing helped reduce noise and improved the model’s ability to generalize to unseen test data.

Based on all experiments, this tuned Multinomial Naive Bayes model is the best-performing classifier.

The original Multinomial Naive Bayes model achieved **85.3%** accuracy. After additional tuning of preprocessing settings and classifier parameters, the accuracy increased to **90.5%** in the second attempt, and then improved further to **91.3%** in the third attempt.

The highest result obtained in all experiments was **91.3%**, achieved by the tuned **Multinomial Naive Bayes** model.

To address this, a simple approach we might try is to throw out the bad data points, and train our model on only the "clean" data.

Come up with a simple heuristic to identify data points containing HTML, and filter out the bad data points to create a cleaned training set.

```python
def is_bad_data(review: str) -> bool:
    return '<' in review and '>' in review
```

### Exercise 3: Identifying Bad Data Points

To improve data quality, the placeholder function was replaced with a real rule to detect problematic reviews.

From Exercise 2, many bad rows were found to contain HTML tags such as `<br>`, `</tr>`, and `<body>`. These rows also often had noisy text or incorrect labels.

Therefore, a simple heuristic was used: if a review contains the symbols `<` and `>`, it is likely to contain HTML formatting and will be treated as bad data.

This allows the dataset to be cleaned before training the model again.

```python
train_clean = train[~train['review'].map(is_bad_data)]
```

### 🔵 Tuning Multinomial Naive Bayes (Second Attempt)

Since the default Multinomial Naive Bayes model performed well, I applied further tuning to both the preprocessing steps and classifier settings in order to improve accuracy.

In this experiment:
- **ngram_range=(1,2)** allows the model to learn both single words and two-word phrases.
- **use_idf=False** removes inverse document frequency weighting.
- **sublinear_tf=True** scales repeated word counts more smoothly.
- **alpha=5.0** adjusts the smoothing parameter of the Naive Bayes classifier.

This remains a model-centric approach because the dataset is unchanged while the model configuration is improved.

```python
nb_clf_tuned = Pipeline([
    ('vect', CountVectorizer(ngram_range=(1,2))),
    ('tfidf', TfidfTransformer(use_idf=False, sublinear_tf=True)),
    ('clf', MultinomialNB(alpha=5.0)),
])

_ = nb_clf_tuned.fit(train['review'], train['label'])

evaluate(nb_clf_tuned)
```

```python
def is_bad_data(review: str) -> bool:
    return '<' in review and '>' in review
```

### Exercise 3: Identifying Bad Data Points

To improve data quality, the placeholder function was replaced with a real rule to detect problematic reviews.

From Exercise 2, many bad rows were found to contain HTML tags such as `<br>`, `</tr>`, and `<body>`. These rows also often had noisy text or incorrect labels.

Therefore, a simple heuristic was used: if a review contains the symbols `<` and `>`, it is likely to contain HTML formatting and will be treated as bad data.

This allows the dataset to be cleaned before training the model again.

```python
train_clean = train[~train['review'].map(is_bad_data)]
```
